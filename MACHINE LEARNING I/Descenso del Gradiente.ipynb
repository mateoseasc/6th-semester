```python
# DESCENSO DEL GRADIENTE

import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.metrics import mean_squared_error, mean_absolute_error

n = 200
X = np.linspace(-10,10,n)
y = X - 0.3 + 1.3 * np.random.randn(n)

#Datos con relacion lineal 
plt.scatter(X,y,color='blue',edgecolors='k',label='Datos')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Datos con relacion lineal')
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt

theta0 = 0.0
theta1 = 0.0
alpha = 0.01
iterations = 30
n = len(X)

losses = []  # aquí guardamos la pérdida por iteración

for _ in range(iterations):
    dtheta0 = 0.0
    dtheta1 = 0.0
    loss = 0.0

    for i in range(n):
        y_hat = theta0 + theta1 * X[i]
        error = y_hat - y[i]

        dtheta0 += error
        dtheta1 += error * X[i]
        loss += error**2

    # promedio
    dtheta0 /= n
    dtheta1 /= n
    loss = loss / (2 * n)

    # guardar pérdida
    losses.append(loss)

    # actualización
    theta0 -= alpha * dtheta0
    theta1 -= alpha * dtheta1

# Gráfica de la función de pérdida
plt.figure()
plt.plot(losses)
plt.xlabel("Iteraciones")
plt.ylabel("Función de pérdida (J)")
plt.title("Convergencia del Descenso del Gradiente")
plt.show()
```
