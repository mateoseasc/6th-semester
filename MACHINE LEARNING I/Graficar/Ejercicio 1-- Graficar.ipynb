```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
df.head()
```

```python
# Grafica de Pastel (Datos nominales) % de estudiantes por carrera 
carreras = df['Carrera'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(carreras, labels=carreras.index, autopct='%1.1f%%', startangle=140)
plt.title('Porcentaje de Estudiantes por Carrera')
```

```python
#Grafica de Barras (Datos discretos u Ordinales), Numero de estudiantes por carrera
plt.figure(figsize=(10,6))
plt.bar(carreras.index, carreras.values, color='purple', edgecolor='black', alpha=0.7)
plt.xlabel('Carreras')
plt.ylabel('Número de Estudiantes')
plt.title('Número de Estudiantes por Carrera')
plt.xticks(rotation=45)
```

```python
#Grafico de Histograma (Datos Continuos) Calificaciones

Calificaciones = df['Promedio']
plt.figure(figsize=(10,6))
plt.hist(Calificaciones, bins=10, color='green', edgecolor='black', alpha=0.7)
plt.xlabel('Rango de Calificaciones')
plt.ylabel('Número de Estudiantes')
plt.title('Distribución de Calificaciones de Estudiantes')
print('Total de Estudiantes:', len(Calificaciones))
```

```python
#Diagrama de Dispersion (Datos pareados) Calificacion vs hrs de estudio

hrs_estudio = df['Horas_estudio']

plt.figure(figsize=(10,6))
plt.scatter(hrs_estudio, Calificaciones, color='black', alpha=0.6)
plt.xlabel('Horas de Estudio por Semana')
plt.ylabel('Calificaciones')
plt.title('Relación entre Horas de Estudio y Calificaciones')
plt.grid(True)
plt.show()
```

```python
#Diagrama de cajas (Boxplot) Calificaciones por Carrera
plt.figure(figsize=(10,6))
df.boxplot(column='Promedio', by='Carrera', grid=False, patch_artist=True,
              boxprops=dict(facecolor='lightblue', color='blue'),
              medianprops=dict(color='red'))
```
