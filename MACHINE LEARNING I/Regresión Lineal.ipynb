```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error

np.random.seed(42)

n = 200
# Crea 200 puntos distribuidos entre -10 y 10
X = np.linspace(-10, 10, n)

# Datos con relacion lineal
# Nota: La imagen muestra un signo menos (-) después de la X, aunque en regresión 
# a veces se usa multiplicación (*). He puesto (-) para ser fiel a la imagen.
y = X - 0.3 + 1.3 * np.random.randn(n)

# Configuración de la gráfica
plt.scatter(X, y, color = 'cornflowerblue', edgecolors='k', label = 'Datos')
plt.xlabel('X')
plt.ylabel('y')
plt.title("Datos")
plt.grid()
plt.show()
```

```python
print (15*"-", "minimos cuadrados", 15*"-")
x_mean = X.mean()
y_mean = y.mean()

numerador = np.sum((X - x_mean) * (y - y_mean))
denominador = np.sum((X - x_mean)**2)

b1 = numerador / denominador
b0 = y_mean - b1 * x_mean
y_pred = b0 + b1 * X

print("b (interseccion):", b0)
print("m (pendiente):", b1)
print("MSE:", mean_squared_error(y, y_pred))
print("MAE:", mean_absolute_error(y, y_pred))

plt.scatter(X, y = y, color = 'blue', edgecolors='black', label = 'Datos')
plt.plot(X, y_pred, color = 'red', label = 'Modelo Regresion lineal')
plt.xlabel('X')
plt.ylabel('y')
plt.legend(loc ='best')
plt.title('Regresion Lineal - Minimos Cuadrados')
plt.show()
```
