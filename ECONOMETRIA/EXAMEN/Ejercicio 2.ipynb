```python
import pandas as pd

# 1. Parámetros del modelo (extraídos de la imagen)
intercepto = 10
b1, b2, b3 = 0.60, -1.20, -0.80

# 2. Interpretación de coeficientes (Ceteris Paribus)
print("--- 1. Interpretación de Coeficientes ---")
print(f"B1 ({b1}): Por cada unidad de Ingreso, el Consumo sube {b1} unidades.")
print(f"B2 ({b2}): Por cada unidad de Inflación, el Consumo baja {abs(b2)} unidades.")
print(f"B3 ({b3}): Por cada unidad de Tasa de Interés, el Consumo baja {abs(b3)} unidades.")
print("(Todo lo anterior manteniendo las demás variables constantes)\n")

# 3. Cálculo de consumo estimado
x1_val, x2_val, x3_val = 86, 4.6, 6.5
y_hat = intercepto + (b1 * x1_val) + (b2 * x2_val) + (b3 * x3_val)

print("--- 2. Consumo Estimado ---")
print(f"Para X1={x1_val}, X2={x2_val}, X3={x3_val}:")
print(f"Y_estimado = {y_hat:.2f}\n")

# 4. Cálculo de VIF y Multicolinealidad
# R^2 auxiliares dados en la imagen
r2_aux = {'X1': 0.94, 'X2': 0.95, 'X3': 0.96}

print("--- 3. Análisis de VIF ---")
for var, r2 in r2_aux.items():
    vif = 1 / (1 - r2)
    status = "Grave" if vif > 10 else "Aceptable"
    print(f"VIF de {var}: {vif:.2f} -> Multicolinealidad: {status}")

# 5. Acción Correctiva
print("\n--- 4. Acción Correctiva Propuesta ---")
print("Dado que los VIF son altísimos (> 10), hay que eliminar una variable.")
print("Propuesta: Eliminar X3 (Tasa de Interés) por tener el VIF más alto (25.0).")
```
