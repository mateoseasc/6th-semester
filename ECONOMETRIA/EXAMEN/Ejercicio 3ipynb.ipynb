```python
import numpy as np

e = np.array([0.40, 0.35, 0.30, 0.25, 0.20, 0.10, 0.05, 0.00, -0.05, -0.10])

diff_e = np.diff(e)
num = np.sum(diff_e**2)
den = np.sum(e**2)

dw = num / den

print(f"Estadistico DW: {dw:.4f}")
```

El estadístico Durbin–Watson calculado es aproximadamente **DW ≈ 0.052**. Como este valor es muy cercano a 0 (y muy inferior a 2), indica **autocorrelación positiva fuerte**. Además, el test de Breusch–Godfrey reporta un p-value de 0.003 (< 0.05), por lo que se rechaza la hipótesis de no autocorrelación. Ambos resultados confirman autocorrelación de primer orden.

Dos consecuencias: (1) los errores estándar están mal estimados, por lo que los t y p-values no son confiables; (2) aunque MCO sigue siendo insesgado, deja de ser eficiente (pierde la propiedad BLUE).

Dos correcciones: (1) estructural: incluir dinámica, por ejemplo un rezago de la variable dependiente o modelar un AR(1); (2) econométrica: usar GLS (Cochrane–Orcutt o Prais–Winsten) o errores estándar Newey–West.
