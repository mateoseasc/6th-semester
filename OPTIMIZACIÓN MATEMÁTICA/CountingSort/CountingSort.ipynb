# COUNTING SORT


## V1.0   Normal Counting Sort

```python
def countSort(arr):
    output = [0 for i in range((len(arr)))]
    count = [0 for i in range(max(arr)+1)]

    for i in arr:
        count[i] += 1

    for i in range(1,len(count)):
        count[i] += count[i-1]

    for i in range(len(arr)):
        output[count[arr[i]]-1] = arr[i]
        count[arr[i]] -= 1

    return output

arr = [1,2,5,7,1,10,9]

lista_ordenada = countSort(arr)
print(lista_ordenada)
```

## V2.0 Implemented Version Of Counting Sort With Negative Numbers

```python
def CountSortNegatives(arr):
    if not arr:  # Check if the array is empty
        return []

    min_val = min(arr)  # Find minimum value (uses internal comparisons)
    max_val = max(arr)  # Find maximum value (uses internal comparisons)

    output = [0 for i in range(len(arr))]  # Create array same size as input
    count = [0 for i in range(max_val - min_val + 1)]  # Create count array for each number between min and max 
    # count array size is max - min + 1 to cover all possible values including negatives

    for i in arr:  # Go through input array and count occurrences
        count[i - min_val] += 1  # Shift negative numbers to start at index 0

    for i in range(1, len(count)):  # Make count array cumulative
        count[i] += count[i-1]

    for i in range(len(arr)):  # Place elements in output array
        output[count[arr[i] - min_val] - 1] = arr[i]
        count[arr[i] - min_val] -= 1  # Decrease count for next duplicate if any

    return output

arr = [-4, -100, 5, -19, 24, 73, 42]

lista_ordenada = CountSortNegatives(arr)
print(lista_ordenada)
```
