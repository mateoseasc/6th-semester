Make a python script that computes the gradient and the hessian of the following functions:




# FUNCTIONS: 

```python
import numpy as np

def gradient(f, x, h=1e-5):
    """Función para calcular el gradiente usando la definición de
    la derivada númerica
    """
    n = len(x) #Dimension del espacio (en nuestro caso es 2)
    grad = np.zeros(n) #Inicializamos el vector del gradiente en cero

    """
    Hacemos un ciclo en donde le vamos sumando un valor muy pequeño a cada
    componente del vector, y luego se lo restamos para ver como varía esa función
    con respecto a cada componente del vector 
    """
    for i in range(n):  
        xf, xb = x.copy(), x.copy()
        xf[i] += h
        xb[i] -= h
        grad[i] = (f(xf) - f(xb)) / (2 * h)
    return grad

def hessian(f, x, h=1e-5):
    """Aplicamos la misma lógica pero para calcular las segundas derivadas
    que corresponden a la matriz Hessiana"""
    n = len(x)
    H = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            xpp, xpm, xmp, xmm = x.copy(), x.copy(), x.copy(), x.copy()
            xpp[i] += h; xpp[j] += h
            xpm[i] += h; xpm[j] -= h
            xmp[i] -= h; xmp[j] += h
            xmm[i] -= h; xmm[j] -= h
            H[i, j] = (f(xpp) - f(xpm) - f(xmp) + f(xmm)) / (4 * h**2)
    return H


x = np.array([1.0, 1.0])  # punto de evaluación
```

 1.

$$
f_1(x_1, x_2) = (a - x_1)^2 + b (x_2 - x_1^2)^2,
\quad a = 1,\; b = 100
$$

```python
a, b = 1, 100
f1 = lambda x: (a - x[0])**2 + b * (x[1] - x[0]**2)**2

print("f1:")
print("  Gradiente:", gradient(f1, x))
print("  Hessiana:\n", hessian(f1, x))
```
